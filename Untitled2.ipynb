```python
from sklearn.model_selection import GroupKFold

folds = GroupKFold(n_splits=6)


preds = []

COLS_TO_USE = [
            i for i in train_test_transformed.columns
            if i not in COLUMNS_TO_REMOVE + ['Transaction_Number_normed', 'subcard_reg_timestamp', 'card1', 'subcard_categorical', 'TransactionDT_dayOfMonth', 'id_31_BROWSER_VERSION']
        ]

print(COLS_TO_USE, len(COLS_TO_USE))


gbms = []

for fold_ind, (train_idx, val_idx) in enumerate(
    folds.split(
        X=train_test_transformed[:train.shape[0]], 
        y=train_test_transformed[:train.shape[0]]['isFraud'], 
        groups=train_test_transformed[:train.shape[0]]['TransactionDT_split'].tolist()
    )
):

    #if fold_ind == 0:
    #    continue

    print('Fold', fold_ind)

    #     pars = {
    #         'num_leaves': 2 ** 7 - 1,
    #         #'min_data_per_group': 100,
    #         'learning_rate': 0.01,
    #         'feature_fraction': 0.7,
    #         'bagging_fraction': 0.7,
    #         'bagging_seed': RANDOM_STATE,
    #         'bagging_freq': 51,
    #         'cat_smooth': 500,
    #         'lambda_l1': 1.0,
    #         'lambda_l2': 3.0,
    #         'max_bin': 100,
    #         'scale_pos_weight': 7.0,
    #         #'max_cat_to_onehot': 10,
    #         'boosting_type': 'gbdt',
    #         'objective': 'binary',
    #         'metric': ['auc'],
    #         'num_threads': -1,
    #         'seed': RANDOM_STATE
    #     }
    
    pars = {'num_leaves': 200,
          'min_child_samples': 40,
          #'min_sum_hessian_in_leaf': 5e-3,
          #'max_bin': 1023,
          #'min_data_in_leaf': 
          #'scale_pos_weight': 2,
          'objective': 'binary',
          'metric': 'auc',
          'max_depth': 13,
          'learning_rate': 0.01,
          "boosting_type": "goss",
          "top_rate": 0.6,
          "other_rate": 0.1,
          "bagging_freq": 0,
          'nthread': -1,
          
          'pos_bagging_fraction': 0.8,
          'neg_bagging_fraction': 0.01,
          "bagging_fraction": 0.5,
          "bagging_seed": 11,
          'reg_alpha': 0.3,
          'reg_lambda': 0.3,
          
          'feature_fraction': 0.9,
          'min_data_per_group': 25,
          'cat_smooth': 500,
          'max_cat_to_onehot': 8,
            'bagging_seed': RANDOM_STATE,
            'seed': RANDOM_STATE
            
          #'categorical_feature': cat_cols
         }

    cat_c = [i for i in CATEGORICAL_FEATURES
        if i in COLS_TO_USE]

    cur_train = train_test_transformed[:train.shape[0]].iloc[train_idx].copy()
    cur_test = train_test_transformed[:train.shape[0]].iloc[val_idx].copy()
    
    TARGET_ENCODE_COLUMNS = []
    
    for TARGET_ENCODE_COLUMN in TARGET_ENCODE_COLUMNS:
    
        cur_train = kfold_target_encoding(cur_train, [TARGET_ENCODE_COLUMN])
        cur_test = apply_target_encoder(cur_train, cur_test, [TARGET_ENCODE_COLUMN])

        cur_train[f'target_encoded_{TARGET_ENCODE_COLUMN}'] = cur_train[f'target_encoded_{TARGET_ENCODE_COLUMN}'].astype('float32')
        cur_test[f'target_encoded_{TARGET_ENCODE_COLUMN}'] = cur_test[f'target_encoded_{TARGET_ENCODE_COLUMN}'].astype('float32')

    
    TE_COLUMNS = [f'target_encoded_{i}' for i in TARGET_ENCODE_COLUMNS]
    
    #cur_train_positive = cur_train[cur_train.isFraud == 1].copy()
    #cur_train_negative = cur_train[cur_train.isFraud == 0].sample(frac=0.2).copy()
    #cur_train = pd.concat([cur_train_positive, cur_train_negative])
    #cur_train = cur_train.sample(frac=.2)

    # create dataset for lightgbm
    lgb_train = lgb.Dataset(cur_train[COLS_TO_USE + TE_COLUMNS], cur_train['isFraud'])
    lgb_eval = lgb.Dataset(cur_test[COLS_TO_USE + TE_COLUMNS], cur_test['isFraud'], reference=lgb_train)

    gbm = lgb.train(
        pars,
        lgb_train,
        num_boost_round=20000,
        valid_sets=(lgb_eval),
        valid_names=('valid',),
        early_stopping_rounds=100,
        feature_name=COLS_TO_USE + TE_COLUMNS,
        categorical_feature=cat_c,
        verbose_eval=100
    )

    gbms.append(gbm)

    te = train_test_transformed[train_test_transformed['TransactionDT_split'] > 17].copy()
    
    for TARGET_ENCODE_COLUMN in TARGET_ENCODE_COLUMNS:
        te = apply_target_encoder(pd.concat([cur_train, cur_test]), te, [TARGET_ENCODE_COLUMN])
        te[f'target_encoded_{TARGET_ENCODE_COLUMN}'] = te[f'target_encoded_{TARGET_ENCODE_COLUMN}'].astype('float32')

    
    preds.append(gbm.predict(te[COLS_TO_USE + TE_COLUMNS], num_iteration=gbm.best_iteration))


print(np.array(np.array([i.best_score['valid']['auc'] for i in gbms])).mean())
```
